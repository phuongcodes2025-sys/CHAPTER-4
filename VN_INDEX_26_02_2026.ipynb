```python
# 1. Khai báo dữ liệu thị trường hôm nay (Dữ liệu giả lập dựa trên tin tức)
market_data = {
    "VNI_Index": 1879.64,
    "Change_Points": 18.73,
    "Gainers": 136,
    "Losers": 181,
    "VIC_Price": 168900
}

# 2. Tính toán tỷ lệ phần trăm tăng trưởng
change_percent = (market_data["Change_Points"] / (market_data["VNI_Index"] - market_data["Change_Points"])) * 100

# 3. Sử dụng cấu trúc điều kiện (If-Else) để nhận định thị trường
print(f"--- BÁO CÁO NHANH VN-INDEX 26/02/2026 ---")
print(f"Chỉ số hiện tại: {market_data['VNI_Index']} ({change_percent:.2f}%)")

if market_data["Gainers"] < market_data["Losers"]:
    status = "Thị trường 'Xanh vỏ đỏ lòng' (Chỉ số tăng nhờ trụ nhưng mã giảm chiếm ưu thế)"
else:
    status = "Thị trường tăng điểm đồng thuận"

print(f"Trạng thái: {status}")

# 4. Kiểm tra sức mạnh của cổ phiếu VIC (Trụ chính hôm nay)
if market_data["VIC_Price"] >= 168900:
    print("Thông báo: Cổ phiếu VIC đang ở vùng giá trần/kỷ lục, là động lực chính của VNI.")
```
