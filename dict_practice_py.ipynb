```python
menu = {"slice": 2.75, "whole": 12.00, "drink": 1.50, "salad": 4.25 }
order = ["slice", "slice", "drink", "salad", "cookie"]

# 1. Build a price breakdown dict of item -> (quantity, line_total)
breakdown = dict()
missing = list()
total = 0
for item in order:
    if item in menu:
        # Corrected logic: Initialize if item is not in breakdown, otherwise update existing entry.
        if item not in breakdown:
            breakdown[item] = (0, 0) # Initialize quantity and total for a new item

        quantity, line_item_total = breakdown[item]
        breakdown[item] = (quantity + 1, line_item_total + menu[item])
    else:
        missing.append(item)
print("Receipt:")
for item in breakdown:
      quantity, line_total = breakdown[item]
      print("%s x%d = %.2f" % (item, quantity, line_total))
      total += line_total
print(f"Missing items: {missing}")
print("Total: %.2f" % (total))

# 2. Ignore unknow items but track them in missing

# 3. Print a receipt summary: each line and the grand total

# Receipt:
# slice x2  = 5.50
# drink x1  = 1.50
# salad x1  = 4.25
# Missing items: ['cookie']
# Total: 11.25
```

```python
# Định nghĩa menu nhà hàng với các món ăn Quảng Ngãi và giá
menu_quang_ngai = {
    "Mì Quảng": 35000,
    "Bún chả cá": 30000,
    "Ram bắp": 25000,
    "Don": 40000,
    "Bánh xèo": 20000,
    "Trà đào": 15000,
    "Coca Cola": 12000
}

# Đơn hàng ví dụ của khách
order_quang_ngai = [
    "Mì Quảng",
    "Mì Quảng",
    "Don",
    "Trà đào",
    "Bún mắm" # Món không có trong menu
]

# 1. Xây dựng từ điển chi tiết đơn hàng: món ăn -> (số lượng, tổng tiền dòng)
breakdown_quang_ngai = dict()
missing_quang_ngai = list()
total_quang_ngai = 0

for item in order_quang_ngai:
    if item in menu_quang_ngai:
        if item not in breakdown_quang_ngai:
            breakdown_quang_ngai[item] = (0, 0) # Khởi tạo số lượng và tổng tiền cho món mới

        quantity, line_item_total = breakdown_quang_ngai[item]
        breakdown_quang_ngai[item] = (quantity + 1, line_item_total + menu_quang_ngai[item])
    else:
        missing_quang_ngai.append(item)

# 3. In tóm tắt hóa đơn: từng dòng và tổng tiền
print("---- HÓA ĐƠN NHÀ HÀNG QUẢNG NGÃI ----")
print("Món ăn 		 Số lượng 	 Thành tiền")
print("-------------------------------------")

for item in breakdown_quang_ngai:
      quantity, line_total = breakdown_quang_ngai[item]
      print(f"{item:<15} x{quantity:<8} = {line_total:,.0f} VNĐ")
      total_quang_ngai += line_total

print("-------------------------------------")
if missing_quang_ngai:
    print(f"* Các món không có trong menu: {missing_quang_ngai}")
print(f"Tổng cộng: {total_quang_ngai:,.0f} VNĐ")
print("-------------------------------------")
```

### Giải thích mã code (Tiếng Việt):

Đoạn mã này mô phỏng quá trình tạo hóa đơn cho một nhà hàng, cụ thể là các món ăn đặc sản Quảng Ngãi.

1.  **`menu_quang_ngai`**: Đây là một từ điển (dictionary) lưu trữ danh sách các món ăn và giá tương ứng của chúng. Bạn có thể dễ dàng thay đổi tên món và giá ở đây.

2.  **`order_quang_ngai`**: Đây là một danh sách (list) chứa các món mà khách hàng đã đặt. Trong ví dụ này, có cả món 'Bún mắm' không có trong menu.

3.  **Vòng lặp xử lý đơn hàng**:
    *   Mã sẽ duyệt qua từng món trong `order_quang_ngai`.
    *   **Nếu món ăn có trong `menu_quang_ngai`**:
        *   Nó kiểm tra xem món ăn đó đã có trong `breakdown_quang_ngai` (chi tiết hóa đơn) chưa. Nếu chưa, nó sẽ khởi tạo mục nhập cho món đó với số lượng 0 và tổng tiền 0.
        *   Sau đó, nó tăng số lượng lên 1 và cộng thêm giá của món ăn vào tổng tiền của dòng (line_item_total) cho món đó.
    *   **Nếu món ăn KHÔNG có trong `menu_quang_ngai`**:
        *   Món ăn đó sẽ được thêm vào danh sách `missing_quang_ngai`.

4.  **In hóa đơn**:
    *   Mã in tiêu đề và định dạng hóa đơn.
    *   Nó duyệt qua từ điển `breakdown_quang_ngai` để in từng món ăn, số lượng và tổng tiền của từng dòng (món). Đồng thời, nó cộng dồn tổng tiền của các món hợp lệ vào `total_quang_ngai`.
    *   Nếu có bất kỳ món nào không có trong menu (`missing_quang_ngai` không rỗng), nó sẽ in ra danh sách các món đó.
    *   Cuối cùng, nó in ra tổng số tiền khách hàng cần thanh toán cho các món hợp lệ.
